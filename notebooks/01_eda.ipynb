```python
import matplotlib.pyplot as plt

# Age distribution
df['age'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Age grouping 
bins = [17, 29, 39, 49, 59, 69, 99]
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)

df[['age', 'age_group']].head(10)
```

```python
df['y'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Target (y) Distribution')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
age_response = df.groupby('age_group', observed=True)['y'].value_counts(normalize=True).unstack() * 100
age_response = age_response.round(1)
age_response
```

```python
age_response['yes'].plot(kind='bar', color='lightgreen')
plt.title('Yes (%) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage saying Yes')
plt.show()
```

```python
job_response = df.groupby('job')['y'].value_counts(normalize=True).unstack() * 100
job_response['yes'].sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Yes (%) by Job Type')
plt.ylabel('Percentage saying Yes')
plt.show()
```

```python
marital_response = df.groupby('marital')['y'].value_counts(normalize=True).unstack() * 100
marital_response['yes'].plot(kind='bar', color='purple')
plt.title('Yes (%) by Marital Status')
plt.ylabel('Percentage saying Yes')
plt.show()
```

```python
edu_response = df.groupby('education')['y'].value_counts(normalize=True).unstack() * 100
edu_response['yes'].plot(kind='bar', color='teal')
plt.title('Yes (%) by Education Level')
plt.ylabel('Percentage saying Yes')
plt.show()
```

```python
# Response rate is highest among retirees and students.

# Younger customers (under 30) are less likely to subscribe.

# Most people contacted were married; single clients say ‘yes’ slightly more often.
#Older clients (especially 60+) show the highest subscription rates, consistent with the retired segment’s strong engagement.
#However, despite being a small group, students also have a very high response rate — possibly because they are more reachable or more open to offers.
#This suggests two promising segments: retirees (large group, high response) and students (small group, very high response).
```

```python
import sqlite3

# create a new SQLite database file
connection = sqlite3.connect("bank_marketing.db")

# save your dataframe to a table named "clients"
df.to_sql("clients", connection, if_exists="replace", index=False)

# close the connection for now
connection.close()
```

```python
# reopen the connection
connection = sqlite3.connect("bank_marketing.db")

# check the first few rows using SQL directly
query = "SELECT * FROM clients LIMIT 5;"
pd.read_sql(query, connection)
```

```python
query = """
SELECT y, COUNT(*) AS n_customers
FROM clients
GROUP BY y;
"""
pd.read_sql(query, connection)
```

```python
query = """
SELECT 
    y,
    ROUND(AVG(age),1) AS avg_age,
    ROUND(AVG(duration),1) AS avg_duration,
    COUNT(*) AS n
FROM clients
GROUP BY y;
"""
pd.read_sql(query, connection)
```

```python
query = """
SELECT 
    job,
    SUM(CASE WHEN y='yes' THEN 1 ELSE 0 END)*100.0/COUNT(*) AS yes_percent,
    COUNT(*) AS total_customers
FROM clients
GROUP BY job
ORDER BY yes_percent DESC;
"""
pd.read_sql(query, connection)
```

```python
query = """
SELECT 
    age_group,
    SUM(CASE WHEN y='yes' THEN 1 ELSE 0 END)*100.0/COUNT(*) AS yes_percent,
    COUNT(*) AS total_customers
FROM clients
GROUP BY age_group
ORDER BY yes_percent DESC;
"""
pd.read_sql(query, connection)
```
